# Mais sobre expressões regulares

```python
import pandas as pd
import numpy as np
import re

data = {
    'name': [
        "  João Silva ", "Maria_Souza", "Pedro-Oliveira", "Ana Costa",
        "Luiz Santos", "Fernanda G. Almeida", "Carlos A. H. Eduardo", "Julia Dias",
        "Rafael Martins", "Beatriz Lima"
    ],
    'state_code': [
        "SP", "RJ", "MG", "ES", "BA ", "rJ", "SPP", "mg",
        "AM", "DF"
    ],
    'state_name': [
        "São Paulo", "Rio de Janeiro", "Minas Gerais", "Espirito Santo",
        "Bahia", "Rio de janeir0", "São pAUlo", "Minas Geraes",
        "Amazonas", "Distrito Federal"
    ],
    'nasc': [
        "1990-01-15", "1985/03/22", "05-10-1992", "1991.07.30",
        "1988/12/01", "02/29/1996", "1999-11-05", "1987-04-18",
        "2000-08-23", "1993-06-10"
    ]
}
df = pd.DataFrame(data)
display(df)

```

## `findall`

```python
df['nasc_digits'] = df['nasc'].str.findall(r'\d+')

display(df[['nasc', 'nasc_digits']].head())
```

```python
df['nasc_digits_1st'] = df['nasc'].str.findall(r'\d+').str[0]

display(df[['nasc', 'nasc_digits_1st']].head())
```

```python
df['name_initials'] = df['name'].str.findall(r'\b[A-Z]\.')
display(df[['name', 'name_initials']])
```

```python
df.loc[6,['name_initials']].to_list()[0][0]
```

## `match`

```python
df['name_starts_with_J'] = df['name'].str.match(r'^J', na=False)
display(df[['name', 'name_starts_with_J']][df['name_starts_with_J'] == True])
```

## Padronizando datas

```python
# Padronizar formato de data para 'YYYY-MM-DD'
def standardize_date(date_str):
    if pd.isna(date_str): # Handle potential NaN values
        return np.nan
    match = re.match(r'(\d{4})[-/.](\d{2})[-/.](\d{2})', date_str) # YYYY-MM-DD or YYYY/MM/DD or YYYY.MM.DD
    if match:
        return f"{match.group(1)}-{match.group(2)}-{match.group(3)}"
    match = re.match(r'(\d{2})[-/.](\d{2})[-/.](\d{4})', date_str) # DD-MM-YYYY or DD/MM/YYYY or DD.MM.YYYY
    if match:
        return f"{match.group(3)}-{match.group(2)}-{match.group(1)}"
    return date_str # Return original if no pattern matches

df['nasc_standardized'] = df['nasc'].apply(standardize_date)
display(df[['nasc', 'nasc_standardized']])
```
